# Importação de bibliotecas

```python
import pandas as pd
```

# Leitura de dataset


```python
df = pd.read_csv('/content/mushroom_cleaned.csv')
```

```python
df
```

```python
df.isnull().sum()
```

# criação de x e y

```python
y = df['class']
x = df.drop(['class'], axis = 1)
```

```python
x
```

```python
y
```

# criação x treino x teste, y treino e y teste

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,test_size = 0.3)
```

```python
x_treino.shape
```

```python
y_treino.shape
```

```python
x_teste.shape
```

```python
y_teste.shape
```

# criação do modelo


```python
from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)
acuracia = modelo.score(x_teste, y_teste)
print('a acuracia é de:',acuracia)
```

# previsao


```python
previsao = modelo.predict(x_teste[40:45])
print('a previsao é de:', previsao)
```
